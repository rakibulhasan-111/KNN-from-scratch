```python
import numpy as np
import pandas as pd
```

```python
def distance(x,X):
  return np.sqrt((x-X)**2)
def KNN(X,Y,x,K=5):
  m = X.shape[0]
  for i in range(m):
    xi = X[i]
    val = []
    dist = distance(x,xi)
    val.append((dist, Y[i]))
  val = sorted(val,key = lambda x:x[0])[:K]
  val = np.asarray(val)
  new_vals = np.unique(val[:,1], return_counts=True)
  index = new_vals[1].argmax()
  output=new_vals[0][index]
  return output
```

```python
def distance (X, Y):
  x1, x2 = X
  y1, y2 = Y

  dx= x1-x2
  dy= y1-y2

  return math.sqrt(dx*dx + dy*dy)
```
